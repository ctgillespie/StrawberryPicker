```python
import numpy as np
import matplotlib.pyplot as plt
import imageio
import xml.etree.ElementTree as ET
import csv
```

```python
def g(X, theta, L, R):
    x = X[0]
    y = X[1]
    z = X[2]
    g = X[3]
    l0 = L[0]
    l1 = L[1]
    l2 = L[2]
    l3 = L[3]
    t0 = theta[0]
    t1 = theta[1]
    t2 = theta[2]
    t3 = theta[3]
    return np.array([x-(R*t3), \
                     y-(l1*np.sin(t0)-l2*np.sin(t0+t1)+l3*np.sin(t0+t1+t2)), \
                     z-(l0+l1*np.cos(np.pi-t0)+l2*np.cos(t0+t1)-l3*np.cos(t0+t1+t2)), \
                     g-(t0+t1+t2)]).transpose()
```

```python
def dg_dtheta(theta, L, R):
    l0 = L[0]
    l1 = L[1]
    l2 = L[2]
    l3 = L[3]
    t0 = theta[0]
    t1 = theta[1]
    t2 = theta[2]
    t3 = theta[3]
    return np.array([[0,0,0,-R], \
                      [-l1*np.cos(t0)+l2*np.cos(t0+t1)-l3*np.cos(t0+t1+t2), \
                       l2*np.cos(t0+t1)-l3*np.cos(t0+t1+t2), \
                       -l3*np.cos(t0+t1+t2), 0], \
                      [-l1*np.sin(np.pi-t0)+l2*np.sin(t0+t1)-l3*np.sin(t0+t1+t2), \
                      l2*np.sin(t0+t1)-l3*np.sin(t0+t1+t2), \
                      -l3*np.sin(t0+t1+t2), 0], \
                      [-1, -1, -1, 0]])
```

```python
def NewtonRaphson(fg, dg, guess, thresh, L, A, R):
    corr = np.linalg.solve(dg(guess, L, R),fg(A, guess, L, R))
    theta = guess - corr
    while (np.linalg.norm(corr) > thresh): 
        corr = np.linalg.solve(dg(theta, L, R),fg(A, theta, L, R))
        theta = theta - corr
    return theta
```

```python
def BuildArm(L,R,theta):
    l0 = L[0]
    l1 = L[1]
    l2 = L[2]
    l3 = L[3]
    t0 = theta[0]
    t1 = theta[1]
    t2 = theta[2]
    t3 = theta[3]
    Z = np.array([l0])
    Y = np.array([0])
    Z = np.append(Z,l0+l1*np.cos(np.pi-t0))
    Y = np.append(Y,l1*np.sin(t0))
    Z = np.append(Z,l0+l1*np.cos(np.pi-t0)+l2*np.cos(t0+t1))
    Y = np.append(Y,l1*np.sin(t0)-l2*np.sin(t0+t1))
    Z = np.append(Z,l0+l1*np.cos(np.pi-t0)+l2*np.cos(t0+t1)-l3*np.cos(t0+t1+t2))
    Y = np.append(Y,l1*np.sin(t0)-l2*np.sin(t0+t1)+l3*np.sin(t0+t1+t2))
    return Z,Y,R*t0
```

```python
class StartPoint():
    def __init__(self, x, y):
        self.x = x
        self.y = y
        self.placed = True
```

```python
class EndPoint():
    def __init__(self, x, y):
        self.x = x
        self.y = y
        self.placed = False
```

```python
tree = ET.parse('smiletest.xml')
root = tree.getroot()
```

```python
StartPoints = []
EndPoints = []
for circle in root[2]:
    if (circle.attrib['style'][6:12]) == "000000":
        EndPoints.append(EndPoint(float(circle.attrib['cx']),float(circle.attrib['cy'])))
    else: 
        StartPoints.append(StartPoint(float(circle.attrib['cx']),float(circle.attrib['cy'])))
```

```python
def myInterp(start, stop, step):
    vals = np.array([])
    if stop > start:
        curr = start
        while (curr < stop):
            vals = np.append(vals, curr)
            curr = curr + step
        vals = np.append(vals, curr)
    else:
        curr = start
        while (curr > stop):
            vals = np.append(vals, curr)
            curr = curr - step
        vals = np.append(vals, curr)
    return vals
```

```python
def extendArray(vect, num):
    while num > 0:
        vect = np.append(vect, vect[-1])
        num = num - 1 
    return vect
```

```python
def genPath(start_x, start_y, end_x, end_y, z_hold, step):
    Xs = myInterp(start_x, end_x, step)
    Ys = myInterp(start_y, end_y, step)
    if (Xs.size < Ys.size):
        Xs = extendArray(Xs, Ys.size-Xs.size)
    else:
        Ys = extendArray(Ys, Xs.size-Ys.size)
    Zs = z_hold*np.ones(Xs.size)
    close = np.ones(Xs.size)
    return Xs, Ys, Zs, close
```

```python
def pick(x, y, z_hold, min_z, step):
    Zs = myInterp(z_hold, min_z, step)
    close = np.zeros(Zs.size)
    close = np.append(close, np.ones(5))
    Zs = np.append(Zs, min_z*np.ones(5))
    Zs = np.append(Zs, myInterp(min_z, z_hold, step))
    close = extendArray(close, Zs.size-close.size)
    Xs = x*np.ones(Zs.size)
    Ys = y*np.ones(Zs.size)
    return Xs, Ys, Zs, close
```

```python
def place(x, y, z_hold, min_z, step):
    Zs = myInterp(z_hold, min_z, step)
    close = np.ones(Zs.size)
    close = np.append(close, np.zeros(5))
    Zs = np.append(Zs, min_z*np.ones(5))
    Zs = np.append(Zs, myInterp(min_z, z_hold, step))
    close = extendArray(close, Zs.size-close.size)
    Xs = x*np.ones(Zs.size)
    Ys = y*np.ones(Zs.size)
    return Xs, Ys, Zs, close
```

```python
def buildPath(StartPoints, EndPoints):
    start_x = 0
    start_y = 0
    step = 5
    z_hold = 50
    min_z = 0
    
    Xs = np.array([])
    Ys = np.array([])
    Zs = np.array([])
    close = np.array([])
    while len(StartPoints) > 0:
        start = StartPoints.pop()
        x_temp, y_temp, z_temp, close_temp = genPath(start_x, start_y, start.x, start.y, z_hold, step)
        Xs = np.append(Xs, x_temp)
        Ys = np.append(Ys, y_temp)
        Zs = np.append(Zs, z_temp)
        close = np.append(close, close_temp)
        start_x = start.x
        start_y = start.y
        x_temp, y_temp, z_temp, close_temp = pick(start_x, start_y, z_hold, min_z, step)
        Xs = np.append(Xs, x_temp)
        Ys = np.append(Ys, y_temp)
        Zs = np.append(Zs, z_temp)
        close = np.append(close, close_temp)
        
        start = EndPoints.pop()
        x_temp, y_temp, z_temp, close_temp = genPath(start_x, start_y, start.x, start.y, z_hold, step)
        Xs = np.append(Xs, x_temp)
        Ys = np.append(Ys, y_temp)
        Zs = np.append(Zs, z_temp)
        close = np.append(close, close_temp)
        start_x = start.x
        start_y = start.y
        x_temp, y_temp, z_temp, close_temp = place(start_x, start_y, z_hold, min_z, step)
        Xs = np.append(Xs, x_temp)
        Ys = np.append(Ys, y_temp)
        Zs = np.append(Zs, z_temp)
        close = np.append(close, close_temp)
        
    return Xs, Ys, Zs, close
```

```python
L = [0] * 4
L[0] = 75
L[1] = 120
L[2] = 100
L[3] = 70
R = 76.2

thresh = 0.01

Xs, Ys, Zs, close = buildPath(StartPoints, EndPoints)
theta = np.array([100*np.pi/180,60*np.pi/180,90*np.pi/180,180*np.pi/180]).transpose()
theta0s = np.array([])
theta1s = np.array([])
theta2s = np.array([])
theta3s = np.array([])

for i in range(Xs.size):
    x = Xs[i]
    y = Ys[i]
    z = Zs[i]
    gamma = 2*np.pi
    A = [x,y,z,gamma]
    
    theta = NewtonRaphson(g, dg_dtheta, theta, thresh, L, A, R)
    theta0s = np.append(theta0s, theta[0])
    theta1s = np.append(theta1s, theta[1])
    theta2s = np.append(theta2s, theta[2])
    theta3s = np.append(theta3s, theta[3])

fig, (ax1, ax2, ax3, ax4) = plt.subplots(4, 1)
ax1.plot(Xs)
ax1.set_ylabel('X')
ax2.plot(Ys)
ax2.set_ylabel('Y')
ax3.plot(Zs)
ax3.set_ylabel('Z')
ax4.plot(close)
ax4.set_ylabel('Grab')

fig, (ax1, ax2, ax3, ax4) = plt.subplots(4, 1)
ax1.plot(theta0s)
ax1.set_ylabel('theta0')
ax2.plot(theta1s)
ax2.set_ylabel('theta1')
ax3.plot(theta2s)
ax3.set_ylabel('theta2')
ax4.plot(theta3s)
ax4.set_ylabel('theta3')
```

```python
with open('motor_commmands.csv','w',encoding='UTF8',newline='') as f:
    writer = csv.writer(f)
    for i in range(len(theta0s)):
        writer.writerow([theta0s[i],theta1s[i],theta2s[i],theta3s[i],close[i]])
```
